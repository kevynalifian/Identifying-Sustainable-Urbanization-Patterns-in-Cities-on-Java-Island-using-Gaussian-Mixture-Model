```python
data["Kab/Kota"] = data["Kab/Kota"].astype("string")
data["Jumlah Kejahatan yang Dilaporkan"] = data["Jumlah Kejahatan yang Dilaporkan"].str.replace(",", "", regex=False).astype("float64")
data["Jumlah UMKM"] = data["Jumlah UMKM"].str.replace(",", "", regex=False).astype("float64")
data["PDRB per Kapita atas Dasar Harga Berlaku (ribu rupiah)"] = data["PDRB per Kapita atas Dasar Harga Berlaku (ribu rupiah)"].str.replace(",", "", regex=False).astype("float64")
data["Panjang Jalan  (km)"] = data["Panjang Jalan  (km)"].str.replace(",", "", regex=False).astype("float64")
data["Jumlah Fasilitas Kesehatan"] = data["Jumlah Fasilitas Kesehatan"].str.replace(",", "", regex=False).astype("float64")
```

```python
import matplotlib.pyplot as plt

numerical_columns = data.columns[1:]  
for col in numerical_columns:
    plt.figure(figsize=(6, 4))
    plt.hist(data[col], bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

```python
for col in numerical_columns:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=data[col])
    plt.title(f'Box Plot of {col}')
    plt.xlabel(col)
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.preprocessing import RobustScaler, StandardScaler, MinMaxScaler, MaxAbsScaler

scaler = MaxAbsScaler()
data_scaled = scaler.fit_transform(data[numerical_columns])
data_scaled_df = pd.DataFrame(data_scaled, columns=numerical_columns)
```

```python
data_scaled_df
```

```python
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

n_components = range(1, 11)

bic = []

for k in n_components:
    gmm = GaussianMixture(n_components=k, random_state=42)
    gmm.fit(data_scaled)  
    bic.append(gmm.bic(data_scaled))

plt.figure(figsize=(10, 6))
plt.plot(n_components, bic, marker='o')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Nilai')
plt.title('Nilai BIC untuk Tiap Jumlah Cluster (GMM)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
bic
```

```python
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(data_scaled)
labels = gmm.predict(data_scaled)
```

```python
data['Hasil Cluster'] = labels
```

```python
from sklearn.decomposition import PCA
import seaborn as sns

pca = PCA(n_components=2)
pca_data = pca.fit_transform(data_scaled)

df_pca = pd.DataFrame(data=pca_data, columns=['PC1', 'PC2'])
df_pca['Cluster'] = labels

centroids_pca = pca.transform(gmm.means_)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='Cluster', palette='Set2', s=100, edgecolor='k')
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], c='red', marker='X', s=180, edgecolor='k', label='Centroid GMM')
plt.title('Visualisasi Hasil Clustering GMM dengan Centroid')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

feature_importance_by_cluster = {}

for cluster_id in sorted(y.unique()):
    # Buat label biner untuk One-vs-Rest
    y_binary = (y == cluster_id).astype(int)

    # Latih Random Forest
    rf = RandomForestClassifier(n_estimators=100, random_state=42)
    rf.fit(X, y_binary)

    # Ambil feature importance
    importances = pd.Series(rf.feature_importances_, index=X.columns)
    top5 = importances.sort_values(ascending=False).head(5)

    feature_importance_by_cluster[f'Klaster {cluster_id}'] = top5

# Tampilkan
for cluster, top_feats in feature_importance_by_cluster.items():
    print(f"\nTop 5 Feature Importance di {cluster}:")
    print(top_feats)
```

### *Interpretasi Hasil Clustering GMM*

Data hasil clustering menggunakan Gaussian Mixture Model (GMM) mengelompokkan kabupaten/kota di Indonesia ke dalam *3 klaster utama* berdasarkan indikator lingkungan, ekonomi, dan sosial. Berikut analisis mendetail untuk setiap klaster:

---

#### *Klaster 0: Kota/Kabupaten dengan Pembangunan Berkelanjutan*
- *Karakteristik Utama*:
  - *Lingkungan*:  
    - Kualitas udara PM2.5 rendah (*≤20 µg/m³), persentase RTH relatif tinggi (5-15%), dan IKLH tinggi (60-70*).  
    - Kejadian banjir/kebakaran rendah (*≤5 kejadian/tahun*).  
  - *Ekonomi*:  
    - PDRB per kapita menengah (*Rp25.000–50.000 ribu), realisasi investasi moderat (Rp0.5–5 triliun*).  
  - *Sosial*:  
    - IPM sedang (*70–75), tingkat pengangguran rendah (4–6%), dan kemiskinan terkendali (≤100 ribu jiwa*).  
- *Contoh*:  
  Kabupaten Pangandaran (PM2.5: 20.4 µg/m³, IKLH: 64.12), Kota Salatiga (IPM: 84.99, PDRB: Rp84.307 ribu).  
- *Interpretasi*:  
  Klaster ini mencerminkan *pembangunan seimbang* dengan fokus pada lingkungan dan kesejahteraan sosial. Cocok sebagai model urbanisasi berkelanjutan.

---

#### *Klaster 1: Kota Metropolitan dengan Pertumbuhan Ekonomi Tinggi dan Risiko Lingkungan*
- *Karakteristik Utama*:
  - *Lingkungan*:  
    - Kualitas udara buruk (PM2.5 *30–50 µg/m³), IKLH rendah (45–55), RTH minim (≤5%*).  
    - Kepadatan penduduk sangat tinggi (>10.000/km²**).  
  - *Ekonomi*:  
    - PDRB per kapita sangat tinggi (*Rp100.000–800.000 ribu), realisasi investasi besar (Rp10–40 triliun*).  
  - *Sosial*:  
    - IPM tinggi (*≥80), tetapi kemiskinan signifikan (≥100 ribu jiwa) dan kesenjangan pendapatan tinggi (GINI **0.35–0.45*).  
- *Contoh*:  
  Jakarta Barat (PM2.5: 53.8 µg/m³, PDRB: Rp237.025 ribu), Surabaya (IKLH: 54.52, Investasi: Rp3.31 triliun).  
- *Interpretasi*:  
  Klaster ini mewakili *urbanisasi tidak berkelanjutan*—pertumbuhan ekonomi cepat, tetapi polusi dan kesenjangan sosial tinggi.

---

#### *Klaster 2: Daerah dengan Tantangan Sosial-Lingkungan Kompleks*
- *Karakteristik Utama*:
  - *Lingkungan*:  
    - Kualitas udara sedang (PM2.5 *20–25 µg/m³), IKLH moderat (55–65), tetapi kejadian banjir tinggi (≥10 kejadian/tahun*).  
  - *Ekonomi*:  
    - PDRB per kapita rendah (*≤Rp50.000 ribu), investasi minim (≤Rp2 triliun*).  
  - *Sosial*:  
    - IPM rendah (*≤70), kemiskinan ekstrem (≥200 ribu jiwa), dan pengangguran tinggi (≥7%*).  
- *Contoh*:  
  Kabupaten Garut (IPM: 68.11, Kemiskinan: 260.5 ribu jiwa), Kabupaten Cirebon (Kejadian banjir: 23 kali/tahun).  
- *Interpretasi*:  
  Klaster ini menghadapi *risiko multidimensi*: lingkungan rentan, ekonomi lemah, dan ketimpangan sosial. Perlu intervensi terpadu.

---

### *Pola dan Rekomendasi Kebijakan*
1. *Klaster 0 (Berkelanjutan)*:  
   - Pertahankan kebijakan lingkungan (e.g., perluasan RTH) dan tingkatkan investasi hijau.  
   - Contoh: Kota Batu (RTH 48%, IPM 78.18).  

2. *Klaster 1 (Tidak Berkelanjutan)*:  
   - Kurangi polusi udara melalui regulasi emisi industri dan transportasi ramah lingkungan.  
   - Tingkatkan akses layanan kesehatan untuk mengurangi kesenjangan.  
   - Contoh: Jakarta Utara (PM2.5: 74.2 µg/m³, GINI 0.54).  

3. *Klaster 2 (Rentan)*:  
   - Perkuat infrastruktur pengendalian banjir dan program pengentasan kemiskinan.  
   - Dorong UMKM berbasis lokal untuk meningkatkan PDRB.  
   - Contoh: Kabupaten Indramayu (Kemiskinan: 214.7 ribu jiwa, UMKM: 13.178 unit).  

---

### *Faktor Pembeda Utama*
- *Lingkungan*: PM2.5, IKLH, dan RTH membedakan klaster 0 (baik) dan 1 (buruk).  
- *Ekonomi*: PDRB per kapita dan investasi memisahkan klaster 1 (ekonomi kuat) dari klaster 0/2.  
- *Sosial*: IPM dan kemiskinan menjadi pembeda utama klaster 2 (sosial lemah).  

### *Kesimpulan*
Data ini menunjukkan *tiga wajah urbanisasi di Indonesia*:  
1. *Berkelanjutan* (Klaster 0): Perlu dipertahankan sebagai model.  
2. *Tidak Berkelanjutan* (Klaster 1): Butuh mitigasi lingkungan dan sosial.  
3. *Rentan* (Klaster 2): Prioritas penanganan multidisiplin.  

Dengan pemahaman ini, pemerintah dapat merancang kebijakan yang lebih tepat sasaran untuk mencapai tujuan pembangunan berkelanjutan (SDGs).
